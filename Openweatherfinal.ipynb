```python
from citipy import citipy as cp
from random import uniform

#To get random float for latitude and longitude
def randfloat():
   return uniform(-180,180), uniform(-90, 90)
points = (randfloat() for x in range(1,2500))

#adding lat and lon into a list of tuple
lat_lon = []
for point in points:
    lat_lon.append(point)


#the * unpack the list or tuple
#to get only unique cities unitl we get 500 cities
cities = []
for i in lat_lon:
    city = cp.nearest_city(*i)
    if city.city_name not in cities and len(cities) < 500:
        cities.append(city.city_name)
    else:
        pass
        
# using set to remove duplicate and get unique city, testing again to see if statement work or not
uniqueciti = set(cities)
uniquecities = list(uniqueciti)
print(f"number of unique city is {len(uniquecities)}.")
print('')
print('')
cities.sort()
print(cities)
```

```python
import json
import requests as req
from datetime import date, datetime
import time

# Save config information
api_key = "d570b78b54d1a33ec03a9055aad07f55"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
counter = 0

#appending data to a list
city_name = []
max_temp = []
lon = []
lat = []
wind_speed = []
max_temp = []
humidity = []
cloud = []
country = []
date = []

for uniquecity in cities[:250]:
    counter += 1
    city = uniquecity
    query_url = url + "appid=" + api_key + "&q=" + city + "&units=" + units
    
    weather_response = req.get(query_url)
    weather_json = weather_response.json()
    
    #printing cities name, number, and url
    print(f"City number is {counter} out of 500.")
    print(weather_json['name'])
    print(query_url)
    
    lon.append(weather_json['coord']['lon'])
    lat.append(weather_json['coord']['lat'])
    city_name.append(weather_json['name'])
    humidity.append(weather_json['main']['humidity'])
    wind_speed.append(weather_json['wind']['speed'])
    max_temp.append(weather_json['main']['temp_max'])
    cloud.append(weather_json['clouds']['all'])
    country.append(weather_json['sys']['country'])
    date.append(datetime.fromtimestamp(int(weather_json["dt"])).strftime('%Y-%m-%d %H:%M:%S'))

#to stop api call for 5mins to reduce the chance of block
time.sleep(5)

for uniquecity in cities[250:]:
    counter += 1
    city = uniquecity
    query_url = url + "appid=" + api_key + "&q=" + city + "&units=" + units
    
    weather_response = req.get(query_url)
    weather_json = weather_response.json()
    
    print(f"City number is {counter} out of 500.")
    print(weather_json['name'])
    print(query_url)
    
    lon.append(weather_json['coord']['lon'])
    lat.append(weather_json['coord']['lat'])
    city_name.append(weather_json['name'])
    humidity.append(weather_json['main']['humidity'])
    wind_speed.append(weather_json['wind']['speed'])
    max_temp.append(weather_json['main']['temp_max'])
    cloud.append(weather_json['clouds']['all'])
    country.append(weather_json['sys']['country'])
    date.append(datetime.fromtimestamp(int(weather_json["dt"])).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
#creating panda dataframe
import pandas as pd
cities_weather = pd.DataFrame({
    "cityname": city_name,
    "temp": max_temp,
    "longtitude": lon,
    "latitude": lat,
    "windspeed": wind_speed,
    "humidity": humidity,
    "cloudiness": cloud,
    "country": country,
    "date_time": date
})
cities_weather.head()
```

```python
#save to csv
cities_weather.to_csv('cities_weather.csv')
```

```python
#Use this to csv so that no need to do api call again
import pandas as pd
cities_w = "cities_weather.csv"

#reading csv
cities_weather = pd.read_csv(cities_w, encoding = 'latin-1')
cities_weather.head()
```

```python
# Temperature (F) vs. Latitude
from datetime import datetime

today_date = datetime.today()

maxtemp = cities_weather['temp']
Lats = cities_weather['latitude']

import matplotlib.pyplot as plt
import numpy as np

x_axis = Lats
y_axis = maxtemp


plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="black",
            s=20, alpha=0.75)
plt.ylim(-100, 150)
plt.xlim(-80, 100)
plt.title(f"City Latitude vs Max Temperature {today_date.strftime('%d-%m-%Y')}")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")


plt.savefig("temp_lat.png")
plt.show()

print('Max Temperature is higher in cities that are closer to the equator and lower in cities that are farther away from equator.')
```

```python
# Humidity (%) vs. Latitude
Humidity = cities_weather['humidity']

x_axis = Lats
y_axis = Humidity

plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="black",
            s=20, alpha=0.75)
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.title(f"City Latitude vs. Humidity {today_date.strftime('%d-%m-%Y')}")
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")


plt.savefig("lat_hum.png")
plt.show()

print('Cities that are closer to the equator has higher humidity than the cities that are farther away')
```

```python
# Cloudiness (%) vs. Latitude
clouds = cities_weather['cloudiness']

x_axis = Lats
y_axis = clouds

plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="black",
            s=20, alpha=0.75)
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.title(f"City Latitude vs Cloudiness {today_date.strftime('%d-%m-%Y')}")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")


plt.savefig("lat_cloud.png")
plt.show()
print('There is no concernable trends between cloudiness and cities location from the equator')
```

```python
# Wind Speed (mph) vs. Latitude

wind_sp = cities_weather['windspeed']

x_axis = Lats
y_axis = wind_sp

plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="black",
            s=20, alpha=0.75)
plt.ylim(-5, 40)
plt.xlim(-80, 100)
plt.title(f"City Latitude vs Wind Speed {today_date.strftime('%d-%m-%Y')}")
plt.xlabel("Latitude")
plt.ylabel("wind speed(mph)")


plt.savefig("lat_wind.png")
plt.show()
print('There seems to be no relationship between windspeed and cities location relative to the equator')
```
